### Confronto tra Voti Utenti e Medie Hotel

### Analisi per Fasce di Punteggio (Binning)
Per comprendere meglio la distribuzione dei giudizi lasciati dai singoli utenti, applichiamo la tecnica del **binning** alla colonna `Reviewer_Score`. Suddividendo i voti in 5 intervalli (bins), possiamo osservare la frequenza delle recensioni per ogni fascia di punteggio. Questo passaggio è fondamentale per identificare eventuali sbilanciamenti nel comportamento dei recensori.


```python
fasce_voti_reviewer = hr['Reviewer_Score'].value_counts(bins=5).sort_index()
print(fasce_voti_reviewer)
```

### Distribuzione dei Punteggi Medi degli Hotel
Eseguiamo la medesima analisi di discretizzazione sulla colonna `Average_Score`. Mentre il punteggio precedente riguarda la singola esperienza dell'utente, questa variabile rappresenta la valutazione globale dell'hotel. Confrontare queste due distribuzioni ci aiuta a capire se i singoli recensori tendono a essere più severi o più generosi rispetto alla media storica della struttura.


```python
fasce_voti_average = hr['Average_Score'].value_counts(bins=5).sort_index()
print(fasce_voti_average)
```


Dall'analisi delle frequenze emerge una forte asimmetria negativa:
1. **Reviewer_Score:** La maggior parte degli utenti (oltre 293.000) assegna voti nell'eccellenza, tra **8.5 e 10**. I voti gravemente insufficienti (sotto il 4) sono una netta minoranza.

### Creazione della Working Copy
Prima di procedere con le operazioni di manipolazione, creiamo una copia del DataFrame originale chiamata `hr1`. Questa è una **best practice** che permette di lavorare sui dati in sicurezza, mantenendo il dataset originale (`hr`) integro e disponibile per eventuali confronti o ripristini.

### Conversione Temporale
Le date sono inizialmente caricate come stringhe (oggetti). Per poter eseguire analisi temporali, trasformiamo la colonna `Review_Date` in oggetti **Datetime**. Successivamente, estraiamo l'**Anno** e il **Mese** in colonne dedicate. 
Questa operazione è fondamentale per studiare la stagionalità delle recensioni e per popolare correttamente lo schema relazionale del database.


```python
hr1['Review_Date'] = p.to_datetime(hr1['Review_Date'])

hr1['Review_Year'] = hr1['Review_Date'].dt.year
hr1['Review_Month'] = hr1['Review_Date'].dt.month

hr1.dtypes
```

### Creazione di variabili derivate:

#### Trasformazione della colonna 'days_since_review'
Ho estratto il valore numerico dalla colonna `days_since_review` per convertirlo in un numero intero (`int`). 

**Perché è stato fatto:**
1. **Pulizia dati:** Il formato originale era testuale (es. "125 days"). Rimuovendo il testo, rendo il dato elaborabile matematicamente.
2. **Analisi dell'anzianità:** Questo valore rappresenta quanti giorni sono passati dal momento della recensione alla data di estrazione del dataset. 
3. **Utilità in SQL:** Trasformare questo campo in un numero permette di eseguire query statistiche (es. media dei giorni, ordinamento cronologico) che non sarebbero possibili su un campo di testo.


```python
hr1['Days_Since_Numeric'] = hr1['days_since_review'].str.extract(r'(\d+)').astype(int) 
```

#### Creazione della variabile derivata 'Total_Review_Length'
Ho creato una nuova colonna che calcola la lunghezza complessiva della recensione sommando il numero di caratteri della parte positiva e di quella negativa.

**Perché è stato fatto:**
1. **Analisi del coinvolgimento:** Permette di misurare quanto l'utente è stato prolisso. In genere, recensioni più lunghe indicano un coinvolgimento maggiore (sia esso positivo o negativo).
2. **Metriche di qualità:** Questa variabile può essere usata per filtrare recensioni troppo brevi o per studiare se esiste una correlazione tra la lunghezza del testo e il punteggio finale (es. "gli utenti insoddisfatti scrivono di più?").
3. **Ottimizzazione SQL:** Avere un valore numerico pre-calcolato facilita le analisi statistiche nel database senza dover processare i campi di testo ogni volta.

```python
hr1['Total_Review_Length'] = hr1['Positive_Review'].str.len() + hr1['Negative_Review'].str.len() 
print(hr1['Total_Review_Length'])
```
