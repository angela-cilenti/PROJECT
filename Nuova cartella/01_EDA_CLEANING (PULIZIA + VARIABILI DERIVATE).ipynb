## Analisi della Distribuzione dei Punteggi
In questa sezione esploriamo come si distribuiscono i voti dei singoli recensori (`Reviewer_Score`) rispetto alla media calcolata per gli hotel (`Average_Score`).


```python
distribuzione = hr[['Reviewer_Score', 'Average_Score']].describe()
print(distribuzione)
```

#### Osservazioni 
* **Volume dati:** Il dataset analizza oltre **515.000** recensioni.
* **Tendenza centrale:** La media dei voti è molto alta (circa **8.4** per entrambi), indicando un set di dati composto prevalentemente da hotel di buona qualità.
* **Variabilità:** Lo scarto quadratico medio (`std`) di `Reviewer_Score` (1.63) è molto più alto di quello di `Average_Score` (0.54). Ciò significa che i singoli utenti hanno opinioni molto varie, mentre le medie degli hotel tendono a somigliarsi molto tra loro.


### Analisi per Fasce di Punteggio (Binning)

```python
fasce_voti_reviewer = hr['Reviewer_Score'].value_counts(bins=5).sort_index()
print(fasce_voti_reviewer)
```

```python
fasce_voti_average = hr['Average_Score'].value_counts(bins=5).sort_index()
print(fasce_voti_average)
```


Dall'analisi delle frequenze emerge una forte asimmetria negativa:
1. **Reviewer_Score:** La maggior parte degli utenti (oltre 293.000) assegna voti nell'eccellenza, tra **8.5 e 10**. I voti gravemente insufficienti (sotto il 4) sono una netta minoranza.
2. **Average_Score:** La quasi totalità degli hotel si concentra nella fascia **7.9 - 8.8**. È estremamente raro trovare in questo dataset hotel con una media globale inferiore a 6.

# Data Cleaning

Procedo innanzitutto a crearmi una copia del DataFrame, così da poterlo manipolare mantenendo intatto il file originale.

```python
hr1=hr.copy()
```

Individuo e rimuovo eventuali righe identiche presenti nel dataset. Questa operazione è fondamentale per evitare che recensioni ripetute alterino i risultati delle analisi statistiche.

```python
hr1.drop_duplicates(inplace=True)
```

Elimino le righe che contengono valori nulli. In questo specifico dataset, i valori mancanti si trovano solitamente nelle coordinate geografiche (latitudine e longitudine). Pulire il dataset dai valori nulli garantisce la coerenza dei calcoli successivi.

```python
hr1.dropna(inplace=True)
```

Utilizzo il metodo `info()` per controllare le dimensioni finali del DataFrame dopo la pulizia e verificare che i tipi di dato siano corretti per ogni colonna.

```python
hr1.info()
```

### Conversione Temporale: 

Trasformo le date in oggetti Datetime ed estraggo Anno e Mese per analisi stagionali.

```python
hr1['Review_Date'] = p.to_datetime(hr1['Review_Date'])

hr1['Review_Year'] = hr1['Review_Date'].dt.year
hr1['Review_Month'] = hr1['Review_Date'].dt.month

hr1.dtypes
```

```python
hr1.info()
```

### Creazione di variabili derivate:

Creo nuove variabili a partire da quelle originali, come la lunghezza del testo e i giorni numerici per migliorare l'analisi SQL.

```python
hr1['Days_Since_Numeric'] = hr1['days_since_review'].str.extract(r'(\d+)').astype(int) # 1. Durata del soggiorno (estraendo il numero dal testo in 'Days_since_review')

hr1['Total_Review_Length'] = hr1['Positive_Review'].str.len() + hr1['Negative_Review'].str.len() # 2. Lunghezza della recensione (sommando i caratteri di recensioni positive e negative)
```

### Export: 

Salvataggio del dataset finale pulito per l'importazione in SQL.

```python
hr1.to_csv('hotel_reviews_clean.csv', index=False)
print("File salvato!")
```
